# ðŸ“± Mobile OS Market Share Visualization

In this notebook, we visualize the market share of mobile operating systems using two chart types:
- A **pie chart** to show proportional distribution
- A **horizontal bar chart** to compare values side-by-side

Weâ€™ll then reflect on which chart is easier to interpret.


```python
# Import necessary library
import matplotlib.pyplot as plt

# Define data
labels = ['Android', 'iOS', 'Others']
market_share = [72, 27, 1]
colors = ['#34A853', '#007AFF', '#AAAAAA']  # Android green, iOS blue, neutral gray
```

## ðŸ¥§ Pie Chart: Market Share Distribution

This chart shows how the mobile OS market is divided proportionally. Itâ€™s useful for getting a quick sense of dominance.


```python
# pie chart
plt.figure(figsize=(6,6))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor' : 'white'})
plt.title('Mobile OS Market Share (Pie Chart)')
plt.axis('equal')  # Ensuring pie is a complete circle
```

## ðŸ“Š Horizontal Bar Chart: Market Share Comparison

This chart allows for easier comparison of exact values, especially helpful when one category is much smaller than the others.


```python
plt.figure(figsize=(8,4))
plt.barh(labels, market_share, color=colors)
plt.xlabel('Market Share (%)')
plt.ylabel('OS')
plt.title('Mobile OS Market Share (Horizontal Bar Chart)')
for i, v in enumerate(market_share):
    plt.text(v + 1, i, f'{v}%', va='center')  # Add value labels
    plt.xlim(0, 80)
```

## ðŸ§  Observation

- The **pie chart** gives a quick visual sense of proportion.
- The **horizontal bar chart** is easier to interpret and compare values, especially for small segments like "Others".
- For presentations or reports, bar charts are generally more readable and scalable.

> âœ… Conclusion: Horizontal bar charts are often more effective for comparing data clearly.
