# Assignment_02:

* Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years. 
* Then create a horizontal bar chart showing total runs scored in their debut 5 years.

## ðŸ“˜ Setup: Import Libraries and Define Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# plot style
sns.set(style="whitegrid")

# Defining data: Runs over selected years
data = {
    'Year': [2001, 2004, 2007, 2010, 2013],
    'Sachin': [1200, 1100, 980, 1000, 850],
    'Kohli':  [0,    0,    1050, 1210, 1340],
    'Sehwag': [900,  1020, 1100, 950,  700]
}

df = pd.DataFrame(data)
```

## ðŸ“Š Chart 1: Side-by-Side Bar Chart (Runs Over Selected Years)

```python
# Melting DF
df_melted = df.melt(id_vars = 'Year', var_name = 'Player', value_name = 'Runs')

# plot
plt.figure(figsize=(10,6))
sns.barplot(data=df_melted, x='Year', y='Runs', hue='Player', palette='viridis')
plt.title('Runs comparison over selected Years')
plt.ylabel("Runs Scored")
plt.xlabel('Year')
plt.legend(title='Player')
plt.tight_layout()
```

## ðŸ“‰ Chart 2: Horizontal Bar Chart (Total Runs in Debut 5 Years)

```python
debut_runs = {
    'Sachin' : sum([600, 800, 750, 900, 850]),
    'Kohli' : sum([0, 0, 500, 800, 950]),
    'Sehwag' : sum([400, 700, 850, 920, 880])
}

debut_df = pd.DataFrame(list(debut_runs.items()), columns=['Player', 'Total Runs'])

# Sorting total runs for better visuals
debut_df = debut_df.sort_values(by='Total Runs', ascending=True)

# Plot
plt.figure(figsize=(8,5))
custom_palette = {
    'Sachin': '#F94144',
    'Kohli': '#577590',
    'Sehwag': '#F9C74F'
}

sns.barplot(
    data=debut_df,
    x='Total Runs',
    y='Player',
    hue='Player',
    palette=custom_palette,
    dodge=False,
    legend=False
) 


plt.title('Total Runs in Debut 5 Years')
plt.xlabel('Runs Scored')
plt.ylabel('Player')
plt.tight_layout()
```
